```python
import pandas as pd
score_AMPSCZ = pd.read_csv('../data-test/AMPSCZ/rawdata/derivatives/scores.tsv', sep="\t")

score_AMPSCZ["score"] = score_AMPSCZ["T1w"].replace('-',-1).fillna(-1).astype(int)
score_AMPSCZ = score_AMPSCZ[score_AMPSCZ["score"]!=-1]
score_AMPSCZ["qc_issue"] = (score_AMPSCZ["score"] >= 0) & (score_AMPSCZ['score']<3)
score_AMPSCZ["bids_name"] = score_AMPSCZ["sub-*/ses-*_gs"]
amp = score_AMPSCZ[["bids_name","qc_issue" , "score" ]].copy()
amp["dataset"]="AMPSCZ"
```

```python
com = score_AMPSCZ[["bids_name","qc_issue" , "score",'T1w comment' ]]
com
```

```python
com["motion"] = com['T1w comment'].str.contains("ringing") | com['T1w comment'].str.contains("motion")
com
```

```python
com.loc[:,"motion_mag"] = com.loc[:,"motion"].astype(int)
com.loc[ com['score']<3, "motion_mag"] +=1
com.to_csv("../data-test/AMPSCZ/rawdata/derivatives/motion_scores.csv")
com
```
